## **Demandas de ML**

### ¿Cuales son los valores promedio de alquiler por tipo de propiedad?

Vamos agrupar el conjunto de datos para saber el promedio por propiedad utilizando la siguiente funcion.

```python
datos.groupby('Tipo')['Valor'].mean()
```

Ahora se organizan los datos desde el mas economico hasta el más costoso.

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Vamos a crear una variable para el comando y despues vamos hacer una grafica.

```python
tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
tipo_precio.plot(kind='barh', figsize=(12,8), color='blue')
```

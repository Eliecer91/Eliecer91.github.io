# Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

Ahora vamos a crear un diccionario

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

Vamos a importar la libreria de matplotlib

```python
import matplotlib.pyplot as plt
```

Agregamos el tamaño de la figura, la informacion de cada 5 años, el titulo y los nombres de los ejes X,Y.

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigrantes de Colombia a Canada')
plt.xlabel('Años')
plt.ylabel('Número de Inmigrantes')
```

Ahora vamos a almacenar el grafico en una variable.

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada')
ax.set_xlabel('Años')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

Realicemos la misma operacion con otro gráfico.

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes de Colombia a Canada')
axs[0].set_xlabel('Años')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Distribución de Inmigrantes')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()
plt.show()
```

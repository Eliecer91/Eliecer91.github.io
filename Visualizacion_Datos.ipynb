# Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

Ahora vamos a crear un diccionario

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

## Matplotlib

Vamos a importar la libreria de matplotlib

```python
import matplotlib.pyplot as plt
```

Agregamos el tamaño de la figura, la informacion de cada 5 años, el titulo y los nombres de los ejes X,Y.

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigrantes de Colombia a Canada')
plt.xlabel('Años')
plt.ylabel('Número de Inmigrantes')
```

Ahora vamos a almacenar el grafico en una variable.

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada')
ax.set_xlabel('Años')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

Realicemos la misma operacion con otro gráfico.

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes de Colombia a Canada')
axs[0].set_xlabel('Años')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Distribución de Inmigrantes')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()
plt.show()
```

Podemos mostrar la descripcion de los datos

```python
datos_col.describe()
```

Vamos analizar los datos de Argentina, Brazil y Perú

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigrantes de los 4 mayores paises suramericanos\nhacia a Canada 1980 - 2013')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[1,0].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Años')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

hacemos unas modificaciones al grafico para que sea mas visual.

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'], lw=2.5,color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia a Canada', fontsize=18)
ax.set_xlabel('Años', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
plt.show()
```

Ahora vamos a crear una variable para los paises de Suramerica.

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

Vamos graficar los valores de cada pais, primero creamos una variable con los colores.

```python
colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Mejoremos el anterior grafico para que sea mas entendible

```python
colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Vamos a organizar el grafico

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Enfoquemos a Colombia para cambiar el color

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y',which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamerica: Colombia es el segundo pais con mas migrantes\nhacia Canada durante el periodo 1980 hasta 2013',
            loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes', fontsize=13)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j), color='black', fontsize=9, ha='left', va='center')

plt.show()
```

Ahora veremos el metodo para exportar los graficos

## Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

Generamos una variable con los 10 paises con mas inmigrantes en el mundo.

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

Ahora su respectivo grafico

```python
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set(title='Top 10 paises con mayor migracion hacia Canada\n1980 - 2013',
       xlabel='Numero de Inmigrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set_title('Top 10 paises con mayor migracion hacia Canada\n1980 - 2013',
              loc='left',fontsize=16)
ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```
